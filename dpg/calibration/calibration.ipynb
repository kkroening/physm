```python
import pandas as pd
import functools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize
```

```python
train = pd.read_csv('calibration_train.csv')
train.head()
```

```python
plt.plot(train.z, train.w)
```

```python
plt.plot(train.z, (train.u - 960/2.) * train.z / 960.)
```

```python
#fig, ax = plt.subplots()
plt.plot(train.z, train.w)
plt.plot(train.z, np.sqrt((train.z.values + 12.5) * 0.08), 'r')
#ax = df.plot(ax=ax, kind='line', x='distance', y='guess')
#plt.show()
```

```python
clip_z = train.z[train.z < -2]
estimate_u = 1000 / -clip_z + 960/2

plt.plot(train.z, train.u, label='actual')
plt.plot(clip_z, estimate_u, label='estimate')
plt.legend()
```

```python
clip_z = train.z[train.z < -2]
estimate_v = 1000 / -clip_z + 540/2

plt.plot(train.z, train.v, label='actual')
plt.plot(clip_z, estimate_v, label='estimate')
plt.legend()
```
