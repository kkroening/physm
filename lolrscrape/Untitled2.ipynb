```python
from matplotlib import pyplot as plt
import pandas as pd

def load_data(filename):
    data = pd.read_csv('../../caliper/{}'.format(filename))
    data = data[data['duration'] < 0.005]
    data['code_prefix'] = data['code'].apply(lambda x: x[:-1])
    return data
    
def get_stats(data):
    return data.groupby('code_prefix')['duration'].aggregate(('mean', 'std'))
```

```python
data2 = load_data('log2.csv')
get_stats(data2)
```

```python
data3 = load_data('log3.csv')
get_stats(data3)
```
